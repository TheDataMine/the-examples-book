# Problem 1: Preprocess and Visualize COVID Map

```python
#Load Data
shapefile = "./UScounties/UScounties.shp"
map_df = gpd.read_file(shapefile)
map_df["FIPS"] = [(lambda x: int(x))(x) for x in map_df["FIPS"].values.tolist()]
map_df = map_df[(map_df["STATE_NAME"] != "Alaska") & (map_df["STATE_NAME"] != "Hawaii")].reset_index(drop=True)
map_df["STATE_FIPS"] = [(lambda x: int(x))(x) for x in map_df["STATE_FIPS"].values.tolist()]
map_df["CNTY_FIPS"] = [(lambda x: int(x))(x) for x in map_df["CNTY_FIPS"].values.tolist()]
```

```python
#Read in CSV data from NYT
df = pd.read_csv("./covid_us_counties.csv")
df = df[df["fips"].notna()]
df["fips"] = [(lambda x: int(x))(x) for x in df["fips"].values.tolist()]
df.head()
```

```python
#Calculate daily case increase column
def calculate_daily_incrase(df, filter_list, select_cols, group_by_cols, group_by, index_column="date", fillna=0):
    """
    Input:
      - df: Pandas DataFrame to calculate daily case or death increase
      - filter_list: List of lists: 
          [[`column` to filter, filter `operator`, filter `criteria`]]
            ([["state", "==", "Michigan"], ["column_name", "==", "criteria"]])
      - select_cols: List of columns to select: [column, column, ...]
      - group_by_cols: List of columns to group by: [column, column, ...]
      - group_by: Pansas group by function as string
      - index_column: Column to use to reset index for daily increase calculation (deafults to "date")
      - fillna: What to use to fill any `nan` values (defaults to 0)
    
    Output:
      - Pandas DataFrame with daily increase in desired matric
    """
    
    #Filter DataFrame------
    #Establish Master String
    master_string = "df["

    #Assemble Filtering
    filters = []
    for filter_item in filter_list:
        #Unpack Values
        column = filter_item[0]
        operator = filter_item[1]
        criteria = filter_item[2]

        #Compile Filter
        if type(criteria) == str:
            filter_string = f"(df['{column}'] {operator} '{criteria}')"
        else:
            filter_string = f"(df['{column}'] {operator} {criteria})"

        #Assemble String
        filters.append(filter_string)

    #Handle Muliple Filters
    if len(filters) > 1:
        filter_string = " & ".join(filters)
    else:
        filter_string = "".join(filters)

    #Add Filtering to Master String
    master_string = master_string + filter_string + "]"

    #Add Select to Master String
    master_string = master_string + f"[{select_cols}]"

    #Add GroupBy Columns and Function to Master String
    master_string = master_string + f".groupby({group_by_cols})" + f".{group_by}()"

    #Filter the DataFrame
    new_df = eval(master_string)
    new_df.reset_index(level=0, inplace=True)
    
    #Calculate Daily Increase
    daily_increase_df = new_df.set_index(index_column).diff()
    daily_increase_df.reset_index(level=0, inplace=True)
    daily_increase_df = daily_increase_df.fillna(fillna)
    
    return daily_increase_df
```

```python
#Get list of FIPS to loop through
fips = df["fips"].unique().tolist()
```

```python
#Empty list for storing dataframes
dfs = []

#Loop through FIPS
for fip in tqdm(fips):
    #FIPS temp df
    _fips_df = df[df["fips"] == fip].reset_index(drop=True).sort_values(by="date", ascending=True)

    _fips_df["daily_increase"] = calculate_daily_incrase(df = _fips_df, 
                                                         filter_list = [["fips", "==", fip]], 
                                                         select_cols = ["date", "cases"], 
                                                         group_by_cols = ["date"], 
                                                         group_by = "sum")["cases"]
    
    #Handle Negatives (https://github.com/nytimes/covid-19-data/issues/456)
    _fips_df["daily_increase"][_fips_df["daily_increase"] < 0] = 0
    
    #Calculate 7-day SMA
    _fips_df["seven_day_moving_avg_new_cases"] = _fips_df.iloc[:,6].rolling(window=7).mean()

    #Append to list
    dfs.append(_fips_df)
```

```python
#Combine lists into master df
df_with_increase = pd.concat(dfs)
```

```python
df_with_increase.head()
```

```python
#JOIN via FIPS Code
merged = map_df.set_index("FIPS").join(df_with_increase.set_index("fips"))

merged.head()
```

```python
#Population data
pop_data = pd.read_csv("./co-est2019-alldata.csv")

#Alter data types to int
pop_data["STATE"] = [(lambda x: int(x))(x) for x in pop_data["STATE"].values.tolist()]
pop_data["COUNTY"] = [(lambda x: int(x))(x) for x in pop_data["COUNTY"].values.tolist()]

#Keep Pop columns
pop_data= pop_data[["STATE", "COUNTY", "CENSUS2010POP"]]
pop_data.head()
```

```python
#Merge population and merged data
merged_pop = pd.merge(merged, pop_data, left_on=['STATE_FIPS','CNTY_FIPS'], right_on = ['STATE','COUNTY'])

#Per 100,000
merged_pop["daily_increase_per_hundred_thousand"] = (merged_pop["seven_day_moving_avg_new_cases"]/merged_pop["CENSUS2010POP"]) * 100000

#Show
merged_pop.head()
```

```python
#Summary Stats
merged_pop.describe()
```

```python
#Keep only target date rows
target_date = "2021-01-06"
target_date_df = merged_pop[merged_pop["date"] == target_date].reset_index(drop=True)
```

```python
#Visualize map
feature = "daily_increase_per_hundred_thousand"

#Set the range for the choropleth
vmin, vmax = 0, 40.144378

#Create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(20, 12))

#Create map
target_date_df.plot(column=feature, cmap="Blues", linewidth=0.8, ax=ax, edgecolor="0.8")
```

```python
#Remove the axis
ax.axis("off")

#Add a title
ax.set_title(f"U.S. County COVID-19 7-day Moving Average Per 100,000 on {target_date}", fontdict={"fontsize": "25", "fontweight" : "3"})

#Create an annotation for the data source
ax.annotate("Source: NYT, 2021", 
            xy=(0.1, .08), 
            xycoords="figure fraction", 
            horizontalalignment="left", 
            verticalalignment="top", 
            fontsize=12, 
            color="#555555")

#I use the dark theme, so I will set my background to white
fig.set_facecolor("white")
```

```python
#Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap="Blues", norm=plt.Normalize(vmin=vmin, vmax=20))

#Empty array for the data range
sm._A = []

#Add the colorbar to the figure
cbar = fig.colorbar(sm)

fig
```
