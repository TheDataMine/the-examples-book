```python
experience = [1,2,3,4,5,6,7,8]

tennis_player_salary = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]

basketball_player_salary = [9000, 12800, 15725, 18000, 19790, 20100, 25700,32000 ]
```

```python
plt.style.use('tableau-colorblind10')

plt.bar(experience,tennis_player_salary,color="r", label= "Tennis Players")
plt.plot(experience,basketball_player_salary, color="g",label= "Basketball Players" )

plt.title("Salary of Tennis Players and Basketball Players by their experiences")
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.legend()
plt.grid(False)

plt.show()
```

```python
#We can also plot multiple bar plots.

plt.style.use("fivethirtyeight")

plt.bar(experience,tennis_player_salary, color="g",linewidth=3,label= "Tennis Players" )
plt.bar(experience,basketball_player_salary,color="r",linewidth=3, label= "Basketball Players")

plt.title("Salary of Tennis Players and Basketball Players by their experiences")
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.legend()
plt.grid(False)

plt.show()
```

```python
plt.style.use("fivethirtyeight")

width = 0.4

experience_indexes = np.arange(len(experience))
plt.bar(experience_indexes - width,tennis_player_salary, color="g",width=width,linewidth=3,label= "Tennis Players" )
plt.bar(experience_indexes+width,basketball_player_salary,color="r",linewidth=3,width=width, label= "Basketball Players")

plt.title("Salary of Data Scientists and Software Engineers by their experiences")
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.legend()
plt.grid(False)

plt.show()
```

```python
#As you can see, x axis does not represent true values of experiences, so we can solve it with xticks() method.

plt.style.use("fivethirtyeight")

width = 0.25

plt.bar(experience_indexes - width, tennis_player_salary, color="g",width=width,linewidth=3,label= "Tennis Players" )
plt.bar(experience_indexes+width,basketball_player_salary,color="r",linewidth=3,width=width, label= "Basketball Players")

plt.title("Salary of Tennis Players and Basketball Players by their experiences")
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.xticks(ticks=experience_indexes, labels=experience)

plt.legend()
plt.grid(True)

plt.show()
```

Stack Plots

```python
days = [1,2,3,4,5,6]

sleep = [6,7,5,8,6,7]
drinking_water =   [2,2,1,2,1,1]
work =  [5,7,10,8,6,9]
exercise=  [3,3,0,1,3,2]
```

```python
plt.style.use("fivethirtyeight")

plt.plot([],[],color='green', label='sleep', linewidth=3)
plt.plot([],[],color='blue', label='drinking_water', linewidth=3)
plt.plot([],[],color='red', label='work', linewidth=3)
plt.plot([],[],color='orange', label='play', linewidth=3)

plt.stackplot(days, sleep, drinking_water, work, exercise, colors=['green','blue','red','orange'])

plt.xlabel('days')
plt.ylabel('activities')
plt.title('6 DAY ROUTINE STACK PLOT EXAMPLE')
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()
```

Scatterplot

```python
first_exam_grades = [89, 90, 70, 89, 100, 80, 90, 100, 80, 34]
second_exam_grades = [30, 29, 49, 48, 100, 48, 38, 45, 20, 30]
```

```python
plt.title("Exam Grades Scatter plot")

plt.scatter(first_exam_grades,second_exam_grades)

plt.tight_layout()
plt.xlabel("First Exam Grades")
plt.ylabel("Second Exam Grades")
plt.grid(True)
plt.show()
```

```python
#We can also plot two different plots.

first_exam_grades = [89, 90, 70, 89, 100, 80, 90, 100, 80, 34]
first_study_hours = [6,8,3,9,9,1,4,2,2,5]
second_exam_grades = [30, 29, 49, 48, 100, 48, 38, 45, 20, 30]
second_study_hours = [2,7,1,5,3,3,2,6,3,2]
```

```python
plt.title("Exam Grades Scatter plot")

plt.scatter(first_exam_grades,first_study_hours,s=100,color="green",marker="x")
plt.scatter(second_exam_grades,second_study_hours,s=100,color="red")

plt.tight_layout()
plt.xlabel("Exam Grades")
plt.ylabel("Study Hours")

plt.grid(True)
plt.show()
```

```python
#You can also add features and colormaps.

first_exam_grades = [89, 90, 70, 89, 100, 80, 90, 100, 80, 34]
second_exam_grades = [30, 29, 49, 48, 100, 48, 38, 45, 20, 30]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,538]
```

```python
plt.title("Exam Grades Scatter plot")

plt.scatter(first_exam_grades,second_exam_grades,s=sizes,c=colors,cmap="Blues",edgecolor="black")

cbar = plt.colorbar()
cbar.set_label("Exam Grades")
plt.tight_layout()
plt.xlabel("First Exam Grades")
plt.ylabel("Second Exam Grades")

plt.grid(True)
plt.show()
```

Time Series

```python
dates = [
    datetime(2021, 3, 10),
    datetime(2021, 3, 13),
    datetime(2021, 3, 14),
    datetime(2021, 3, 15),
    datetime(2021, 3, 16),
    datetime(2021, 3, 17),
    datetime(2021, 3, 18),
    datetime(2021, 3, 19)
]

values = [0,3,4,7,5,3,5,6]
```

```python
plt.title("Time Series")

plt.plot_date(dates, values,linestyle="solid",marker = 'o',ms = 20, mfc = 'r',c="b" )

plt.xticks(rotation='vertical')
plt.xlabel("Dates")
plt.ylabel("Values")
plt.grid(False)

plt.show() 
```

Heat Map \\

Matplotlib's imshow() or heatmap() function makes production of such plots particularly easy. matplotlib.pyplot.pcolormesh() is an alternative function.

```python
data = np.random.random(( 6 , 6 )) 
data
```

```python
plt.imshow( data , cmap = 'autumn' ) 
  
plt.title( "2-D Heat Map" ) 
plt.show() 
```

```python
sns.heatmap( data , linewidth = 0.5 , cmap = 'coolwarm' ) 
  
plt.title( "2-D Heat Map" ) 
plt.show()
```
